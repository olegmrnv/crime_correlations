```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
import csv

from config import census_api_key
from pprint import pprint
```

```python
# Save config information
url = 'https://api.census.gov/data/2017/pep/population?get='

# Build query URL
query_url = url + "POP,GEONAME&for=COUNTY:109&in=state:51&DATE=3,4,5,6,7,8&key=" + census_api_key
response = requests.get(query_url)
response_json = response.json()
print(response_json)
```

```python
# Convert to DataFrame
louisa_df = pd.DataFrame(response_json)

# String replace index
new_header = louisa_df.iloc[0]
louisa_df = louisa_df[1:]
louisa_df.columns = new_header
louisa_df
```

```python
# Rename headers
louisa_df = louisa_df.rename(columns={'POP': 'Population',
                                      'GEONAME': 'County, State',
                                      'DATE': 'Year',
                                      'state': 'State',
                                      'county': 'County'})
louisa_df
```

```python
# Replace values
louisa_df['Year'] = louisa_df['Year'].replace({'3':'2010', '4':'2011', '5':'2012', '6':'2013', '7':'2014', '8':'2015'})
louisa_df["Year"] = pd.to_numeric(louisa_df["Year"])
louisa_df["Population"] = pd.to_numeric(louisa_df["Population"])
louisa_df
```

```python
# Remove columns
louisa_df.drop(columns=['State', 'County'])
```

```python
# Plot the graph
plot = louisa_df.plot(x='Year', y='Population', kind='line', title='Population 2010-2015 Louisa County', legend=True, color='green')
```

```python
# Save in PNG
fig = plot.get_figure()
fig.savefig('Population_2010-2015_Louisa_County.png')
```
