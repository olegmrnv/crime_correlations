```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
import csv

from config import census_api_key
from pprint import pprint
```

```python
# Save config information
url = 'https://api.census.gov/data/2000/pep/int_population?get='

# Build query URL
query_url = url + "POP,GEONAME&for=COUNTY:109&in=state:51&DATE=6,7,8,9,10,11&key=" + census_api_key
response = requests.get(query_url)
response_json = response.json()
print(response_json)
```

```python
# Convert to DataFrame
louisa_df = pd.DataFrame(response_json)

# String replace index
new_header = louisa_df.iloc[0]
louisa_df = louisa_df[1:]
louisa_df.columns = new_header
louisa_df
```

```python
louisa_df.info()
```

```python
# Rename headers
louisa_df = louisa_df.rename(columns={'POP': 'Population',
                                      'GEONAME': 'County_State',
                                      'DATE': 'Year',
                                      'state': 'State',
                                      'county': 'County'})
louisa_df
```

```python
# Replace values
louisa_df['Year'] = louisa_df['Year'].replace({'6':'2004', '7':'2005', '8':'2006', '9':'2007', '10':'2008', '11':'2009'})
louisa_df
```

```python
# Remove columns
louisa_df.drop(columns=['State', 'County'])
```
