PART 2 OLEG

```python
# Base url to pull list of police stations in TX
# Data and AIP key obtained from https://crime-data-explorer.fr.cloud.gov/api

base_url = "https://api.usa.gov/crime/fbi/sapi/api/agencies/byStateAbbr/raw/TX?api_key="
url = base_url + crime_api_key

# Pulling data from FBI Crime
TX_police_stations = requests.get(url).json()
```

```python
# Reviewing received data

pprint (TX_police_stations)
```

```python
# Removing header and keeping dictionary with police stations info
TX_police_stations = TX_police_stations["TX"]
```

```python
# Converting dictionaty to DataFrame
TX_police_stations_pd = pd.DataFrame(TX_police_stations)
TX_police_stations_pd.head()
```

```python
# Switching rows and columns to have table we could easily read and manipulate

TX_police_stations_pd = TX_police_stations_pd.transpose()
TX_police_stations_pd.head()
```

```python
# Since we analyzing  hurricane Ike we will keep only polise stations located in Galveston county
# Saving DataFerame into csv File for future refferals

galveston_police_stations_pd = TX_police_stations_pd.loc[TX_police_stations_pd["county_name"] == "GALVESTON", :]
galveston_police_stations_pd.to_csv("../data/cleaned_data/galveston_police_stations.csv")
```

```python
# Removing warnings appearing during large data manipulation

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading incident-based data in Texas from csv file for 2008 (since hurricane Ike occurred in 2008) 
# and searching for data reported by Galveston police stations only

incident_based_data_TX_2008 = pd.read_csv("../data/nibrs_incident_TX_2008.csv")
incident_based_data_TX_2008.head()
```

```python
# Extracting police station name and getting a list of unique polise stations who reported data in TX

police_stations_reported_data = incident_based_data_TX_2008["ddocname"].str[8:17]
police_stations_reported_data.unique()
```

```python
# Getting the list of police stations in Galveston county who reported data in 2008
galveston_polise_station_name = set(galveston_police_stations_pd.index.values.tolist()) & set(police_stations_reported_data.unique())
```

```python
# Turned out only one police sttaions in Galveston reported incident-based data to FBI system in 2008
galveston_polise_station_name
```

```python
# Extracting station name from SET element and assigning it to new short variable GPS (Galveston Police Station)
GPS = galveston_polise_station_name.pop()
```

```python
# Iterating through each data file, keeping only incident_date and filtering by police station name (GPS)
# each row will represent one offence reported by Galveston Police Station
# we have data files from 2005 till 2010 (3 years before Ike and 2 years after)

incident_based_data = {}
incident_based_data_filtered = {}
years = [2005, 2006, 2007, 2008, 2009, 2010]

for year in years:
    print ("Working on " + str(year) +" year")
    incident_based_data[year] = pd.read_csv("../data/nibrs_incident_TX_" + str(year) + ".csv")
    incident_based_data[year] = incident_based_data[year][["incident_date", "ddocname"]]
    incident_based_data[year]["incident_date"] = incident_based_data[year]["incident_date"].str[:10]
    incident_based_data[year]["ddocname"] = incident_based_data[year]["ddocname"].str[8:17]
    incident_based_data_filtered[year] = incident_based_data[year].loc[incident_based_data[year]["ddocname"] == GPS]
    incident_based_data_filtered[year].drop(columns=['ddocname'], inplace=True)
    incident_based_data_filtered[year].to_csv("../data/cleaned_data/" + GPS + "_" + str(year) + ".csv", index=False)

print ("Completed successfully")   
```

```python
# Base url to pull list of police stations in VA
# Data and AIP key obtained from https://crime-data-explorer.fr.cloud.gov/api

base_url = "https://api.usa.gov/crime/fbi/sapi/api/agencies/byStateAbbr/raw/VA?api_key="
url = base_url + crime_api_key

# Pulling data from FBI Crime
VA_police_stations = requests.get(url).json()
```

```python
# Reviewing received data

pprint (VA_police_stations)
```

```python
# Removing header and keeping dictionary with police stations info
VA_police_stations = VA_police_stations["VA"]
```

```python
# Converting dictionaty to DataFrame
VA_police_stations_pd = pd.DataFrame(VA_police_stations)
VA_police_stations_pd.head()
```

```python
# Switching rows and columns to have table we could easily read and manipulate

VA_police_stations_pd = VA_police_stations_pd.transpose()
VA_police_stations_pd.head()
```

```python
# Earthquake occured in Virginia in Louisa county so we will be looking for police stations in there
# Saving DataFerame into csv File for future refferals

louisa_police_stations_pd = VA_police_stations_pd.loc[VA_police_stations_pd["county_name"] == "LOUISA", :]
louisa_police_stations_pd.to_csv("../data/cleaned_data/louisa_police_stations.csv")
```

```python
# Reading incident-based data in Virginia from csv file for 2011 (since earthquake occurred in 2011) 
# and searching for data reported by Louisa police stations only

incident_based_data_VA_2011 = pd.read_csv("../data/nibrs_incident_VA_2011.csv")
incident_based_data_VA_2011.head()
```

```python
# Extracting police station name and getting a list of unique polise stations who reported data in VA

police_stations_reported_data = incident_based_data_VA_2011["ddocname"].str[8:17]
```

```python
# Getting the list of police stations in Louisa county who reported data in 2011
louisa_polise_station_name = set(louisa_police_stations_pd.index.values.tolist()) & set(police_stations_reported_data.unique())
```

```python
# Turned out all police sttaions in Louisa reported incident-based data to FBI system in 2011
louisa_polise_station_name
```

```python
# Converting SET element into list

louisa_stations = list(louisa_polise_station_name)
louisa_stations
```

```python
# Iterating through each data file, keeping only incident_date and filtering by police station names
# each row will represent one offence reported by Louisa police stations
# we have data files from 2008 till 2014 (3 years before earthquake and 3 years after)

incident_based_data = {}
incident_based_data_filtered = {}
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014]

for year in years:
    print ("Working on " + str(year) +" year")
    incident_based_data[year] = pd.read_csv("../data/nibrs_incident_VA_" + str(year) + ".csv")
    incident_based_data[year] = incident_based_data[year][["incident_date", "ddocname"]]
    incident_based_data[year]["incident_date"] = incident_based_data[year]["incident_date"].str[:10]
    incident_based_data[year]["ddocname"] = incident_based_data[year]["ddocname"].str[8:17]
    incident_based_data_filtered[year] = incident_based_data[year].loc[incident_based_data[year]["ddocname"].isin(louisa_stations)]
    incident_based_data_filtered[year].drop(columns=['ddocname'], inplace=True)
    incident_based_data_filtered[year].to_csv("../data/cleaned_data/louisa_stations_" + str(year) + ".csv", index=False)

print ("Completed successfully") 
```

```python
# Save config information
gal_url1 = 'https://api.census.gov/data/2000/pep/int_population?get='

# Build query URL
gal_query_url1 = gal_url1 + "POP,GEONAME&for=PLACE:28068&in=state:48&DATE=6,7,8,9,10,11&key=" + census_api_key
gal_response1 = requests.get(gal_query_url1)
gal_response_json1 = gal_response1.json()
print(gal_response_json1)
```

```python
# Convert to DataFrame
galveston_df1 = pd.DataFrame(gal_response_json1)
galveston_df1.to_csv("../data/cleaned_data/gal_pop_2004-2009.csv")

# String replace index
gal_new_header1 = galveston_df1.iloc[0]
galveston_df1 = galveston_df1[1:]
galveston_df1.columns = gal_new_header1
galveston_df1
```

```python
# Rename headers
galveston_df1 = galveston_df1.rename(columns={'POP': 'Population',
                                              'GEONAME': 'City, State',
                                              'DATE': 'Year',
                                              'state': 'State',
                                              'place': 'City'})
galveston_df1
```

```python
# Replace values
galveston_df1['Year'] = galveston_df1['Year'].replace({'6':'2004', '7':'2005', '8':'2006', '9':'2007', '10':'2008', '11':'2009'})
galveston_df1["Year"] = pd.to_numeric(galveston_df1["Year"])
galveston_df1["Population"] = pd.to_numeric(galveston_df1["Population"])
galveston_df1
```

```python
# Remove columns
galveston_df1.drop(columns=['State', 'City'])
```
