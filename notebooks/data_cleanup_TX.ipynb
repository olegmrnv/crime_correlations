```python
import pandas as pd
import json
import requests
from api_keys import crime_api_key
from pprint import pprint
```

```python
# Base url to pull list of police stations in TX
# Data and AIP key obtained from https://crime-data-explorer.fr.cloud.gov/api

base_url = "https://api.usa.gov/crime/fbi/sapi/api/agencies/byStateAbbr/raw/TX?api_key="
url = base_url + crime_api_key

# Pulling data from FBI Crime
TX_police_stations = requests.get(url).json()
```

```python
# Reviewing received data

pprint (TX_police_stations)
```

```python
# Removing header and keeping dictionary with police stations info
TX_police_stations = TX_police_stations["TX"]
```

```python
# Converting dictionaty to DataFrame
TX_police_stations_pd = pd.DataFrame(TX_police_stations)
TX_police_stations_pd.head()
```

```python
# Switching rows and columns to have table we could easily read and manipulate

TX_police_stations_pd = TX_police_stations_pd.transpose()
TX_police_stations_pd.head()
```

```python
# Since we analyzing  hurricane Ike we will keep only polise stations located in Galveston county
# Saving DataFerame into csv File for future refferals

galveston_police_stations_pd = TX_police_stations_pd.loc[TX_police_stations_pd["county_name"] == "GALVESTON", :]
galveston_police_stations_pd.to_csv("../data/cleaned_data/galveston_police_stations.csv")
```

```python
# Removing warnings appearing during large data manipulation

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading incident-based data in Texas from csv file for 2008 (since hurricane Ike occurred in 2008) 
# and searching for data reported by Galveston police stations only

incident_based_data_TX_2008 = pd.read_csv("../data/nibrs_incident_TX_2008.csv")
```

```python
incident_based_data_TX_2008.head()
```

```python
# Extracting police station name and getting a list of unique polise stations who reported data in TX

police_stations_reported_data = incident_based_data_TX_2008["ddocname"].str[8:17]
police_stations_reported_data.unique()
```

```python
# Getting the list of police stations in Galveston county who reported data in 2008
galveston_polise_station_name = set(galveston_police_stations_pd.index.values.tolist()) & set(police_stations_reported_data.unique())
```

```python
# Turned out only one police sttaions in Galveston reported incident-based data to FBI system in 2008
galveston_polise_station_name
```

```python
# Extracting station name from SET element and assigning it to new short variable GPS (Galveston Police Station)
GPS = galveston_polise_station_name.pop()
```

```python
# Iterating through each data file, keeping only incident_date and filtering by police station name (GPS)
# each row will represent one offence reported by Galveston Police Station
# we have data files from 2005 till 2011 (3 years before Ike and 3 years after)

incident_based_data = {}
incident_based_data_filtered = {}
years = [2005, 2006, 2007, 2008, 2009, 2010, 2011]

for year in years:
    print ("Working on " + str(year) +" year")
    incident_based_data[year] = pd.read_csv("../data/nibrs_incident_TX_" + str(year) + ".csv")
    incident_based_data[year] = incident_based_data[year][["incident_date", "ddocname"]]
    incident_based_data[year]["incident_date"] = incident_based_data[year]["incident_date"].str[:10]
    incident_based_data[year]["ddocname"] = incident_based_data[year]["ddocname"].str[8:17]
    incident_based_data_filtered[year] = incident_based_data[year].loc[incident_based_data[year]["ddocname"] == GPS]
    incident_based_data_filtered[year].reset_index(inplace=True, drop=True)
    incident_based_data_filtered[year].drop(columns=['ddocname'], inplace=True)
    incident_based_data_filtered[year].to_csv("../data/cleaned_data/" + GPS + "_" + str(year) + ".csv")

print ("Completed successfully")   
```
